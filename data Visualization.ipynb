```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

sns.set(style="whitegrid")

#read csv file      
df = pd.read_csv('datafile.csv')
df.columns = df.columns.str.strip()
```

### 1. Seasonal Contribution and Percentage Contribution to Total Rice Production (Top 10 Districts)

```python
import pandas as pd
import matplotlib.pyplot as plt

season_prod_cols = ["Kharif_Production", "Rabi_Production", "Summer_Production"]

df["Total_Production"] = df[season_prod_cols].sum(axis=1)
top10 = df.sort_values("Total_Production", ascending=False).head(10)

top10 = top10.set_index("District Name")

top10_pct = top10[season_prod_cols].div(top10[season_prod_cols].sum(axis=1), axis=0) * 100

fig, axes = plt.subplots(2, 1, figsize=(12, 12), sharex=True)

top10[season_prod_cols].plot(
    kind="bar",
    stacked=True,
    color=["#2ca02c", "#1f77b4", "#ff7f0e"], 
    edgecolor="black",
    ax=axes[0]
)
axes[0].set_title("Seasonal Contribution to Total Rice Production (Top 10 Districts)", fontsize=14)
axes[0].set_ylabel("Production (Tonnes)")
axes[0].legend(title="Season", loc="upper right")
axes[0].grid(axis="y", linestyle="--", alpha=0.7)

top10_pct.plot(
    kind="bar",
    stacked=True,
    color=["#2ca02c", "#1f77b4", "#ff7f0e"],
    edgecolor="black",
    ax=axes[1]
)
axes[1].set_title("Percentage Contribution of Each Season (Top 10 Districts)", fontsize=14)
axes[1].set_ylabel("Percentage (%)")
axes[1].grid(axis="y", linestyle="--", alpha=0.7)

# Final formatting
plt.xticks(rotation=45, ha="right")
plt.xlabel("District")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

season_prod_cols = ["Kharif_Production", "Rabi_Production", "Summer_Production"]

df["Total_Production"] = df[season_prod_cols].sum(axis=1)

df_sorted = df.sort_values("Total_Production", ascending=False)

df_sorted = df_sorted.set_index("District Name")

df_pct = df_sorted[season_prod_cols].div(df_sorted[season_prod_cols].sum(axis=1), axis=0) * 100

fig, axes = plt.subplots(2, 1, figsize=(16, 14), sharex=True)

df_sorted[season_prod_cols].plot(
    kind="bar",
    stacked=True,
    color=["#2ca02c", "#1f77b4", "#ff7f0e"], 
    edgecolor="black",
    ax=axes[0],
    alpha=0.8
)
axes[0].set_title("Seasonal Contribution to Total Rice Production (All Districts)", fontsize=16)
axes[0].set_ylabel("Production (Tonnes)", fontsize=12)
axes[0].legend(title="Season", loc="upper right")
axes[0].grid(axis="y", linestyle="--", alpha=0.7)

df_pct.plot(
    kind="bar",
    stacked=True,
    color=["#2ca02c", "#1f77b4", "#ff7f0e"],
    edgecolor="black",
    ax=axes[1],
    alpha=0.8
)
axes[1].set_title("Percentage Contribution of Each Season (All Districts)", fontsize=16)
axes[1].set_ylabel("Percentage (%)", fontsize=12)
axes[1].grid(axis="y", linestyle="--", alpha=0.7)

plt.xticks(rotation=90, ha="right")
plt.xlabel("District", fontsize=12)
plt.tight_layout()
plt.show()
```

### 2.Total Rice Production and Rice Productivity (Yield)  by District (All Seasons)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df["Total_Production"] = (
    df["Kharif_Production"] +
    df["Rabi_Production"] +
    df["Summer_Production"]
)


df_sorted = df.sort_values("Total_Production", ascending=False)

fig, axes = plt.subplots(2, 1, figsize=(14, 12), sharex=True)


sns.barplot(
    data=df_sorted,
    x="District Name",
    y="Total_Production",
    palette="viridis",
    ax=axes[0]
)
axes[0].set_title("Total Rice Production by District (All Seasons)", fontsize=16)
axes[0].set_ylabel("Production (Tonnes)", fontsize=13)
axes[0].set_xlabel("")
axes[0].tick_params(axis='x', rotation=90)
axes[0].grid(axis='y', linestyle='--', alpha=0.6)

sns.barplot(
    data=df_sorted,
    x="District Name",
    y="All Seasons_Yield",
    palette="mako",
    ax=axes[1]
)
axes[1].set_title("Rice Productivity (Yield) by District", fontsize=16)
axes[1].set_ylabel("Yield (kg/hectare)", fontsize=13)
axes[1].set_xlabel("District", fontsize=13)
axes[1].tick_params(axis='x', rotation=90)
axes[1].grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

### 3. Yield vs Area and Yield vs Production (All Seasons)

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.scatterplot(
    data=df,
    x="All Seasons_AreaAfter bund correction factor",
    y="All Seasons_Yield",
    color="royalblue",   
    s=70 
)
plt.xlabel("Total Area (Hectares)")
plt.ylabel("Yield (Kg/Hectare)")
plt.title("Yield vs Area (All Seasons)")

plt.subplot(1, 2, 2)
sns.scatterplot(
    data=df,
    x="All Seasons_Production",
    y="All Seasons_Yield",
    color="darkorange",
    s=70
)
plt.xlabel("Total Production (Tonnes)")
plt.ylabel("Yield (Kg/Hectare)")
plt.title("Yield vs Production (All Seasons)")

plt.tight_layout()
plt.show()
```

### 4. Heatmap  Area, Yield & Production Correlations

```python
plt.figure(figsize=(8,6))
numeric_cols = [
    "Kharif_AreaAfter bund correction factor",
    "Kharif_Yield", "Kharif_Production",
    "Rabi_AreaAfter bund correction factor",
    "Rabi_Yield", "Rabi_Production",
    "Summer_AreaAfter bund correction factor",
    "Summer_Yield", "Summer_Production",
    "All Seasons_AreaAfter bund correction factor",
    "All Seasons_Yield", "All Seasons_Production"
]

corr = df[numeric_cols].corr()

sns.heatmap(corr, cmap="YlGnBu", annot=False)
plt.title("Correlation Between Area, Yield, and Production Metrics", fontsize=14)
plt.tight_layout()
plt.show()
```

### 5.Yield Efficiency Ranking (All Seasons)

```python
plt.figure(figsize=(16,6))
top_yield = df.sort_values("All Seasons_Yield", ascending=False)

sns.barplot(
    data=top_yield,
    x="District Name",
    y="All Seasons_Yield",
    palette="Spectral"
)
plt.xticks(rotation=45)
plt.title("District-wise Rice Yield Ranking (All Seasons)", fontsize=14)
plt.ylabel("Yield (kg/hectare)")
plt.xlabel("District")
plt.tight_layout()
plt.show()
```
